```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import math
```

```python
telecom_df = pd.read_csv('filtered.csv')
```

```python
#combining 6 and 7th month
col_67 = list(set(['_'.join(colm.split('_')[:-1]) for colm in telecom_df.columns if (colm.endswith('_6') or colm.endswith('_7'))]))

ns_cols = 'fb_user','night_pck_user'
max_cols = ['max_rech_data','max_rech_amt']
either_or_cols = ['date_of_last_rech_data','date_of_last_rech']
wavg_cols = [ 'loc_og_t2t_mou','loc_ic_t2m_mou','loc_ic_mou','total_rech_num','monthly_2g', \
             'onnet_mou','count_rech_3g','sachet_3g','vol_3g_mb','roam_ic_mou','loc_og_t2m_mou', \
             'total_rech_amt','loc_og_t2f_mou','std_og_t2m_mou','loc_og_t2c_mou','std_ic_t2f_mou', \
             'loc_ic_t2t_mou', 'total_rech_data','arpu_2g','count_rech_2g','roam_og_mou','sachet_2g', \
             'og_others','loc_og_mou','std_og_mou','ic_others','total_og_mou','std_og_t2t_mou', \
             'arpu_3g','isd_og_mou','vol_2g_mb','isd_ic_mou','std_ic_mou','std_ic_t2m_mou', \
             'total_ic_mou','spl_ic_mou','spl_og_mou','std_ic_t2t_mou','loc_ic_t2f_mou', \
             'offnet_mou', 'monthly_3g', 'std_og_t2f_mou','arpu','last_day_rch_amt','av_rech_amt_data']

```

```python
#wavg collg
def agg_add(col):
    a = 0.7 ; b = 0.3
    telecom_df[col+'_67'] = telecom_df[col+'_6']*b + telecom_df[col+'_7']*a
    telecom_df.drop([col+'_6',col+'_7'], axis=1,inplace=True)

def direct_assignment(col):
    telecom_df['date_of_last_rech_67'] = telecom_df[col+'_7']
    telecom_df.drop([col+'_6',col+'_7'], axis=1,inplace=True)

def max_assign(col):
    telecom_df['date_of_last_rech_67'] = telecom_df[[col+'_6',col+'_7']].max(axis=1)
    telecom_df.drop([col+'_6',col+'_7'], axis=1,inplace=True)
    
def cat_col_agg(col):
    telecom_df[col+'_67'] = telecom_df[col+'_6']+ telecom_df[col+'_7']*1.5 #more weight for 7th month
    telecom_df.drop([col+'_6',col+'_7'], axis=1,inplace=True)

    
for col in wavg_cols:
    agg_add(col)
    
for col in either_or_cols:
    direct_assignment(col)

for col in max_cols:
    max_assign(col)
    
for col in ns_cols:
    cat_col_agg(col)
```

```python
call_range_ = ['loc','std']
call_type = ['ic','og']
call_mo = ['67','8']
from_to = ['t2t','t2c','t2f','t2m']

for cr in call_range_:
    for ct in call_type:
        for m in call_mo:
            su_ = 0
            str_ = ''
            for ft in from_to:
                pcol = cr+'_'+ct+'_mou_'+m
                scol = cr+'_'+ct+'_'+ft+'_mou_'+m
                if scol in telecom_df.columns:
                    su_+= telecom_df[scol]
                    str_ = str_ + ' + ' + scol
            print(pcol+' = '+str_)
            diff = telecom_df[pcol] -su_
            print(diff.apply(lambda x:np.absolute(x)).describe())
            print('95%tile:     '+str(diff.apply(lambda x:np.absolute(x)).quantile(q=0.95)))
#results of difference distribution show that the max values are outliers 
# Hence we can replace the supersets with the subsets of sums and find ratios
```

```python
call_range_ = ['loc','std','isd','spl','roam']
call_type = ['ic','og']
call_mo = ['67','8']

for m in call_mo:
    for ct in call_type:
        su_ = 0
        str_ = ''
        for cr in call_range_:
            pcol = 'total'+'_'+ct+'_mou_'+m
            scol = cr+'_'+ct+'_mou_'+m
            if scol in telecom_df.columns:
                su_+= telecom_df[scol]
                str_ = str_ + ' + ' + scol
        print(pcol+' = '+str_)
        diff = telecom_df[pcol] -su_
        print(diff.apply(lambda x:np.absolute(x)).describe())
        print('95%tile:     '+str(diff.apply(lambda x:np.absolute(x)).quantile(q=0.95)))
#like the last case the outliers are much ahead of the 75% hence those might be outliers
```
